```python
import numpy as np
import subprocess
```

```python
def reduce_g(num):
    gstr = f"{num:<10.10G}"
    
    if "." not in gstr:
        gstr = f"{num:<10.10E}"
    
    if 'E' in gstr:
        splt = gstr.split('E')
        return 'E'.join([splt[0][:9-len(splt[1])], splt[1]])
    else:
        return gstr
```

```python
for inflection_angle in np.linspace(3, 10, 30):

    ## Card 1 Title
    
    title = "Mach 6"
    axisymmetric = True

    card1 = f" {title:^10.10s} {axisymmetric-1:>2d}"

    ## Card 2 Gas Props
    
    gamma = 1.40
    gas_constant = 1716.563
    compressability_factor = 1.0
    turbulent_recovery_factor = 0.896
    viscosity_constant_C = 2.26968E-8
    viscosity_constant_M = 198.72
    fifth_deg_velocity_distribution_SFOA = 0
    nozzle_end_XBL = 1000

    card2info = [gamma, gas_constant, compressability_factor, turbulent_recovery_factor, 
                 viscosity_constant_C, viscosity_constant_M, fifth_deg_velocity_distribution_SFOA, nozzle_end_XBL]

    card2 = "".join([reduce_g(num) for num in card2info])

    ## Card 3 Upstream
    
#     inflection_angle = 8.67
    ratio_of_throat_radius_to_curvature = 6.0
    mach_num_point_f = 0
    mach_num_point_b = 5.0
    design_mach_num_at_c = 0
    
    
    
    
    inviscid_exit_radius = -1 * 3             # exit radius in inches 
    x_pos_point_a = 0                         # inflection point x position
    distance_source_point_c = 0

    card3info = [inflection_angle, ratio_of_throat_radius_to_curvature, mach_num_point_f,
                mach_num_point_b, design_mach_num_at_c, inviscid_exit_radius, 
                x_pos_point_a, distance_source_point_c]

    card3 = "".join([reduce_g(num) for num in card3info])

    ## Card 4 Resolution
    
    MT = 41
    NT = 21
    IX = 0
    IN = 10
    IQ = 0
    MD = 41
    ND = 49
    NF = -61
    MP = 0
    MQ = 0
    JB = 1
    JX = 0
    JC = 10
    IT = 0
    LR = -21
    NX = 13

    card4info = [MT, NT, IX, IN, IQ, MD, ND, NF, MP, MQ, JB, JX, JC, IT, LR, NX]

    card4 = "".join([f"{num:<5d}" for num in card4info])

    ## Card 5 Smoothing
    
    smoothing_upstream_count = 50
    smoothing_factor_percent = 85
    smoothing_downstream_count = 50

    card5info = [smoothing_upstream_count, smoothing_factor_percent, smoothing_downstream_count]

    card5 = "".join([f"{num:<5d}" for num in card5info])

    ## Card 6 Plenum / heat transfer stuff
    
    
    PPQ = 90                     # stagnation pressure (psia)
    TO = 1030                    # stagnation temp (rankine)
    TWT = 540                    # throat wall temp (rankine)
    TWAT = 540                   # B wall temp (rankine) 
    QFUN = 0
    ALPH = 0
    IHT = 0
    IR = 0
    ID = 1
    LV = 5

    card6info1 = [PPQ, TO, TWT, TWAT, QFUN, ALPH]
    card6info2 = [IHT, IR, ID, LV]

    card6 = "".join([*[reduce_g(num) for num in card6info1], *[f"{num:<5d}" for num in card6info2]])

    ## Card 7 meshing
    
    XST = 1000
    XLOW = 46
    XEND = 172
    XINC = 2
    BJ = 0
    XMID = 0
    XINC2 = 0
    CN = 0

    card7info = [XST, XLOW, XEND, XINC, BJ, XMID, XINC2, CN]

    card7 = "".join([reduce_g(num) for num in card7info])

    gens = [card1, card2, card3, card4, card5, card6, card7]

    with open('input.txt', 'w') as out_file:
        for line in gens:
            out_file.write(line + '\n')
            
    subprocess.call('contur.exe')
    with open('output.txt', 'r') as in_file:
        sim_res = in_file.read()
    print(sim_res.split('\n')[-2:-1])
```
